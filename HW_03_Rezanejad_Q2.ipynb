```python
import pandas as pd
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
print("Raw data frame:")
print(df.head())
```

```python
print("Null columns:")
print(df.isnull().sum())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
print("After filling the null columns:")
print(df.isnull().sum())
```
